# Subgroup Analysis

This notebook reads each model's `test_predictions.csv` and writes model-specific subgroup AUROC files.

Outputs per model:
- `gender_subgroup.csv`
- `age_subgroup.csv`
- `view_subgroup.csv`


```ipython
from pathlib import Path
import sys

PROJECT_ROOT = Path.cwd().resolve()
if PROJECT_ROOT.name == 'notebooks':
    PROJECT_ROOT = PROJECT_ROOT.parent
sys.path.insert(0, str(PROJECT_ROOT))

CHECKPOINT_DIR = PROJECT_ROOT / 'checkpoints'
MODELS = ['densenet', 'efficientnet', 'vit']

print(f'Project root: {PROJECT_ROOT}')
print(f'Checkpoint dir: {CHECKPOINT_DIR}')
```

```ipython
from src.analysis.run_subgroup_analysis import analyze_model

all_outputs = {}
for model_key in MODELS:
    outputs = analyze_model(CHECKPOINT_DIR, model_key)
    all_outputs[model_key] = outputs
    gender_mean = outputs['gender_subgroup.csv'].query("Disease == 'Mean'").iloc[0]
    print(
        f"{model_key}: gender mean M={gender_mean['Male AUROC']:.4f}, "
        f"F={gender_mean['Female AUROC']:.4f}; "
        f"age rows={len(outputs['age_subgroup.csv'])}; "
        f"view rows={len(outputs['view_subgroup.csv'])}"
    )
```

```ipython
for model_key in MODELS:
    print(f'\n[{model_key}] gender_subgroup.csv')
    display(all_outputs[model_key]['gender_subgroup.csv'].tail(5))
    print(f'[{model_key}] age_subgroup.csv')
    display(all_outputs[model_key]['age_subgroup.csv'])
    print(f'[{model_key}] view_subgroup.csv')
    display(all_outputs[model_key]['view_subgroup.csv'])
```
